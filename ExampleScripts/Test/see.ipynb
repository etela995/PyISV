```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
embed = np.load('embed.npy')
reconstruction = np.load('rescaled_outputs.npy')

inputs = np.load('random_inputs.npy')
targets = np.load('random_targets.npy')
```

```python
fig,ax = plt.subplots(ncols=1, nrows=2, figsize=(10,6))
ax1,ax2 = ax.flatten()
idx = np.random.permutation(np.arange(len(inputs)))[0]
ax1.set_title('Index = %.d' % idx)
ax1.plot(inputs[idx], c='crimson', label='Input')
ax1.legend(fontsize = 12)
ax1.set_xticks([])

ax2.plot(targets[idx],c='navy', label='Target')
ax2.plot(reconstruction[idx], c='cyan', label='Reconstruction')
ax2.legend(fontsize = 12)

fig.subplots_adjust(hspace=0.05)
plt.show()
```
